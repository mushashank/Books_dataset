```python
def date_change(value):
    if not pd.isnull(value):
        dt = value[:6]
        if '[' in dt[0]:
            date = dt[1:5]
            return date
        else:
            date = dt[:4]
            return date
        
```

```python
df['Date of Publication'] = df['Date of Publication'].apply(date_change)
```

```python
def date_cln(date):
    try:
        return np.int64(date)
    except:
        return np.nan
```

# Split name that start with 'pp' at ':' and extract name of the place after ':'

```python
def extract_place(place):
    if place.startswith('pp.'):
        try:
            part_plc = place.split(':')[1].strip()
            if ',' in part_plc:
                place = part_plc[:part_plc.find(',')]
            elif ';' in part_plc:
                place = part_plc[:part_plc.find(',')]
            return place
        except:
            return np.nan
    else:
        if 'London' in place:
            place = 'London'
        return place
        
```

```python
df['Place of Publication'] = df['Place of Publication'].apply(extract_place)
```
